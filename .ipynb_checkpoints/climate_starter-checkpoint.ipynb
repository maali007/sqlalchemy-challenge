```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

from flask import Flask, jsonify
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect = True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
conn = engine.connect()
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
mr_date = engine.execute('SELECT max(date) FROM measurement').fetchone()
print(mr_date[0])
```

```python
# Calculate the date one year from the last date in data set.
last12 = (dt.datetime.strptime(mr_date[0], '%Y-%m-%d') - dt.timedelta(days=365)).date()
```

```python
# Perform a query to retrieve the data and precipitation scores
pptn = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= last12).order_by(Measurement.date).all()
#print(pptn)
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
pptn_df = pd.DataFrame(pptn).set_index("date")
#print(pptn_df)
```

```python
# Sort the dataframe by date
pptn_df = pptn_df.sort_values("date")
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
pptn_df.plot(y = "prcp", figsize = (12,12))
plt.title(f"Exploratory Precipitation Analysis: ({(last12)} to {mr_date[0]})")
plt.xlabel("Date")
plt.ylabel("Inches")
plt.xticks(rotation = "45", horizontalalignment = "right")
plt.show()

```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
pptn_df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
stations_count = session.query(Station.id).count()
print("There are {} stations in the dataset.".format(stations_count))
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
stations_active = session.query(Measurement.station,func.count(Measurement.station)).group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
for station in stations_active:
    #print(row)
    # List the stations and the counts in descending order.
    print(f"Station {station[0]} total number of row = {station[1]}. ")
```

```python
# Find the most active station id from the previous query
print(f"The most active station from the list above has the ID {stations_active[0][0]}.")
```

```python
# From the station above, calculate the lowest, highest, and average temperature.
min_tempo = round(session.query(func.min(Measurement.tobs)).filter(Measurement.station == "USC00519281").scalar(),2)
max_tempo = round(session.query(func.max(Measurement.tobs)).filter(Measurement.station == "USC00519281").scalar(),2)
avg_tempo = round(session.query(func.avg(Measurement.tobs)).filter(Measurement.station == "USC00519281").scalar(),2)
print(f"The lowest observed temperature is {min_tempo} degrees Fahrenheit.")
print(f"The average observed temperature is {avg_tempo} degrees Fahrenheit.")
print(f"The highest observed temperature is {max_tempo} degrees Fahrenheit.")
```

```python
# Using the most active station id, query the last 12 months of temperature observation data for this station and plot the results as a histogram

temp_observ = session.query(Measurement.tobs).filter(Measurement.station==stations_active[0][0]).filter(Measurement.date >= last12).order_by(Measurement.date.desc()).all()
temp_observ_df = pd.DataFrame(temp_observ)
temp_observ_df.head()

plt.hist(temp_observ_df['tobs'], bins=12)
plt.ylabel("Frequency")
plt.title(f"Station Analysis ({(last12)} - {mr_date[0]}) for Station { stations_active[0][0]}")
plt.legend(["tobs"])
plt.show()
```
